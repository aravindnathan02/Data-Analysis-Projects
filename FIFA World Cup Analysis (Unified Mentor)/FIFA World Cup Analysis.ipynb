```python
# Data Cleaning
matches = matches.drop(columns=['Datetime', 'Referee', 'Assistant 1', 'Assistant 2'])
players = players.drop(columns=['Coach Name', 'Shirt Number'])
cups = cups.drop(columns=['QualifiedTeams'])
```

```python
matches['Total Goals'] = matches['Home Team Goals'] + matches['Away Team Goals']
```

```python
updated_matches = '/content/drive/My Drive/FIFA/updated_matches.csv'
matches.to_csv(updated_matches, index=False)
updated_players = '/content/drive/My Drive/FIFA/updated_players.csv'
players.to_csv(updated_players, index=False)
updated_cups = '/content/drive/My Drive/FIFA/updated_cups.csv'
cups.to_csv(updated_cups, index=False)
```

```python
# Total no. of world cups
cups.value_counts().sum()
```

```python
# Goals Scored by Home Teams each Year
home_year = matches.groupby('Year')['Home Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=home_year, x='Year', y='Home Team Goals', marker='o')
for x, y in zip(home_year['Year'], home_year['Home Team Goals']):
    plt.annotate(f'{y}', (x, y), textcoords='offset points', xytext=(0, 5), ha='center')
plt.xticks(rotation=90)
plt.title('Goals Scored by Home Teams each Year')
plt.show()
```

```python
# Goals Scored by Home teams
home_goals = matches.groupby('Home Team Name')['Home Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=home_goals.sort_values(by='Home Team Goals', ascending=False), x='Home Team Name', y='Home Team Goals', hue='Home Team Goals')
plt.xticks(rotation=90)
plt.title('Goals Scored by Home Teams')
plt.show()
```

```python
# Goals Scored by Away Teams each Year
away_year = matches.groupby('Year')['Away Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=away_year, x='Year', y='Away Team Goals', marker='o')
for x, y in zip(away_year['Year'], away_year['Away Team Goals']):
    plt.annotate(f'{y}', (x, y), textcoords='offset points', xytext=(0, 5), ha='center')
plt.xticks(rotation=90)
plt.title('Goals Scored by Away Teams each Year')
plt.show()
```

```python
# Goals Scored by Away Teams
away_goals = matches.groupby('Away Team Name')['Away Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=away_goals.sort_values(by='Away Team Goals', ascending=False), x='Away Team Name', y='Away Team Goals', hue='Away Team Goals')
plt.xticks(rotation=90)
plt.title('Goals Scored by Away Teams')
plt.show()
```

```python
# Home Teams Winning World Cups
home_win = cups[cups['Country'] == cups['Winner']]

plt.figure(figsize=(12, 6))
sns.scatterplot(home_win, x='Year', y='Winner')
for x, y in zip(home_win['Year'], home_win['Winner']):
    plt.annotate(f'{x}', (x, y), textcoords='offset points', xytext=(0, 5), ha='center')
plt.xticks(rotation=90)
plt.title('Home Teams Winning World Cups')
plt.show()
```

```python
# Away teams winning world cup
away_win = cups[cups['Country'] != cups['Winner']]

plt.figure(figsize=(12, 6))
sns.scatterplot(away_win, x='Year', y='Winner')
for x, y in zip(away_win['Year'], away_win['Winner']):
    plt.annotate(f'{x}', (x, y), textcoords='offset points', xytext=(0, 5), ha='center')
plt.xticks(rotation=90)
plt.title('Away Teams Winning World Cups')
plt.show()
```

```python
# Goals Scored by Home Team within half time
home_half = matches.groupby('Home Team Name')['Half-time Home Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=home_half.sort_values(by='Home Team Name', ascending=True), x='Home Team Name', y='Half-time Home Goals')
plt.xticks(rotation=90)
plt.title('Goals Scored by Home Team within Half Time')
plt.show()
```

```python
# Goals Scored by Away Team within half time
away_half = matches.groupby('Away Team Name')['Half-time Away Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=away_half.sort_values(by='Away Team Name', ascending=True), x='Away Team Name', y='Half-time Away Goals')
plt.xticks(rotation=90)
plt.title('Goals Scored by Away Team within Half Time')
plt.show()
```

```python
# Goals scored by home teams in different stages of tournament
home_stage = matches.groupby(['Stage'])['Home Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=home_stage.sort_values(by='Home Team Goals', ascending=False), x='Stage', y='Home Team Goals', hue='Home Team Goals', legend='auto')
plt.xticks(rotation=90)
plt.title('Goals Scored by Home Teams in Different Stages of Tournament')
plt.show()
```

```python
# Goals scored by away teams in different stages of tournament
away_stage = matches.groupby(['Stage'])['Away Team Goals'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=away_stage.sort_values(by='Away Team Goals', ascending=False), x='Stage', y='Away Team Goals', hue='Away Team Goals', legend='auto')
plt.xticks(rotation=90)
plt.title('Goals Scored by Away Teams in Different Stages of Tournament')
plt.show()
```

```python
# Historical trends in world cups
# Average goals per match
avg_goals = matches.groupby('Year')['Total Goals'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_goals, x='Year', y='Total Goals', marker='o')
for x, y in zip(avg_goals['Year'], avg_goals['Total Goals']):
    plt.annotate(f'{y:.2f}', (x, y), textcoords='offset points', xytext=(0, 5), ha='center')
plt.xticks(rotation=90)
plt.title('Average Goals per Match')
plt.xlabel('Year')
plt.ylabel('Average Goals per Match')
plt.legend()
plt.show()
```

```python
# Average goals in each stages
goals_stage = matches.groupby('Stage')['Total Goals'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=goals_stage.sort_values(by='Total Goals', ascending=False), x='Stage', y='Total Goals', hue='Total Goals', legend='auto')
plt.xticks(rotation=90)
plt.title('Average Goals in Each Stages')
plt.show()
```

```python
# Total goals scored in each world cup
plt.figure(figsize=(12, 6))
sns.barplot(data=cups, x='Year', y='GoalsScored', hue='GoalsScored')
for p in plt.gca().patches:
    plt.gca().annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.xticks(rotation=90)
plt.title('Total Goals Scored in Each World Cup')
plt.show()
```

```python
# Matches won after extra time
extra_time = matches['Win conditions'].str.contains('extra time', case=False)
matches['Wins after Extra Time'] = extra_time
extra_time = matches.groupby('Year')['Wins after Extra Time'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=extra_time, x='Year', y='Wins after Extra Time')
plt.xticks(rotation=90)
plt.title('Matches Won After Extra Time')
plt.show()
```

```python
# Matches won on penalties
penalties = matches['Win conditions'].str.contains('penalties', case=False)
matches['Wins on Penalties'] = penalties
penalties = matches.groupby('Year')['Wins on Penalties'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=penalties, x='Year', y='Wins on Penalties')
plt.xticks(rotation=90)
plt.title('Matches Won on Penalties')
plt.show()
```

```python
# Top 50 stadiums attendance in terms of Attendance
stadiums = matches.groupby('Stadium')['Attendance'].sum().reset_index()
top_stadiums = stadiums.nlargest(50, 'Attendance')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_stadiums, x='Stadium', y='Attendance', hue='Attendance')
plt.xticks(rotation=90)
plt.title('Top 50 Stadiums')
plt.show()
```

```python
# Number of players who made impact by starting and not starting in the line-up
line_up = players.groupby('Line-up').count().reset_index()

plt.figure(figsize=(12, 6))
plt.pie(data=line_up, x='Event', labels='Line-up', autopct='%1.1f%%')
plt.title('Events caused by players starting and not starting in line-up')
plt.show()
```
